```python
month = range(1, 13)
year = range(2016, 2022)

url = 'https://www.lightreading.com/archives.asp?newsandviews=yes&tag_id=758&piddl_month={}&piddl_year={}'.format(month, year)


def pause():
    programpause = input("Press the <ENTER> key to continue...")


def pause2():
    programpause = input("Ã‡a fait une heure que je dors")


articles_month = []
articles_author = []
articles_date = []
articles_text = []
articles_title = []
articles_final = []
articles_link = []
articles_snippet = []

counter_loop = 0
counter = 0
```

```python
counter = 0
for i in range(2016, 2022, 1):
    for j in range(1, 13, 1): 
        if j < 8 and i == 2016:
            continue
        elif j > 2 and i == 2021:
            continue
        else:
            url = 'https://www.lightreading.com/archives.asp?newsandviews=yes&tag_id=758&piddl_month={}&piddl_year={}'.format(j, i)
            articles_month.append(url)
```

```python
for i in articles_month:
    
    j = requests.get(i)
    soup = BeautifulSoup(j.text, 'html.parser')
    
    k = soup.find_all("p", class_="card-body")
    if k is not None:
        articles_snippet.append(k)
        counter += 1
    else:
        articles_snippet.append("Not_found")
        counter += 1

    if counter == 70:
        print("Sleepin", current_time)
        sleep(3700)
        print("Wakin op", current_time)
        counter = 0


print(articles_snippet[-10])
print(len(articles_snippet))
df_articles_snippet = pd.DataFrame(articles_snippet)
df_articles_snippet.to_csv(r'C:\Users\Kevin\Documents\Python\article_snippet.csv', index=False)

df_test = pd.DataFrame(articles_snippet)
df_test.to_csv(r'C:\Users\Kevin\Documents\Python\article_test.csv', mode='a', index=False)
```

```python
for i in articles_month:
    
    j = requests.get(i)
    soup = BeautifulSoup(j.text, 'html.parser')

    for i in soup.find_all("div", class_="card-content"):
        j = i.a.get("href")
        articles_link.append(j)
        counter += 1

        if counter == 70:
            df_test = pd.DataFrame(articles_final)
            df_test.to_csv(r'C:\Users\Kevin\Documents\Python\article_test.csv', mode='a', index=False)
            print("Sleepin", current_time)
            sleep(3700)
            print("Wakin up", )
            counter = 0


url_append = "https://www.lightreading.com"
articles_final = [url_append + sub for sub in articles_link]

print(articles_final[-10])
print(len(articles_final))
df_articles_final = pd.DataFrame(articles_final)
df_articles_final.to_csv(r'C:\Users\Kevin\Documents\Python\article_final.csv', index=False)

df_test = pd.DataFrame(articles_final)
df_test.to_csv(r'C:\Users\Kevin\Documents\Python\article_test.csv', mode='a', index=False)
```

```python

import csv
with open('article_final.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader : 
        j = requests.get(row[0])
        soup = BeautifulSoup(j.text, 'html.parser')
        counter += 1
        counter_loop += 1
    
        k = soup.find('span', class_='date')
        if k is not None:
            articles_date.append(k)
        else:
            articles_date.append("Not_found")
    
        l = soup.find('span', class_='allcaps')
        if l is not None:
            articles_author.append(l)
        else:
            articles_author.append("Not_found")

        m = soup.find('article')
        if m is not None:
            n = m.get_text()
            articles_text.append(n)
        else:
            articles_text.append("Not_found")
    
        o = soup.find('h1', class_='article-title')
        if o is not None:
            articles_title.append(o)
        else:
            articles_title.append("Not_found")

        if counter == 70:
            data_list = [articles_author, articles_title, articles_date, articles_text, articles_snippet, articles_final]
            df = pd.concat([pd.Series(x) for x in data_list], axis=1)
            df.to_csv(r'C:\Users\Kevin\Documents\Python\article_test.csv', mode='a', index=False)
            print("Sleepin", current_time)
            print(counter_loop)
            sleep(3700)
            print("Wakin up", current_time)
            counter = 0
```
